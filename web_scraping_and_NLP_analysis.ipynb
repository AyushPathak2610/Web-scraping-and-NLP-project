```python
import os

stopwords = set()
stop_folder = 'StopWords'

for filename in os.listdir(stop_folder):
    file_path = os.path.join(stop_folder, filename)
    if os.path.isfile(file_path) and filename.endswith('.txt'):
        with open(file_path, 'r') as file:
            for line in file:
                words = line.split()
                stopwords.update(words)

stopwords = list(stopwords)
stopwords
```

```python
master_folder = 'MasterDictionary'
positive_words = list()
negative_words = list()

for filename in os.listdir(master_folder):
    file_path = os.path.join(master_folder, filename)
    if os.path.isfile(file_path) and filename.endswith('.txt'):
        file_name = os.path.splitext(filename)[0]

        with open(file_path, 'r') as file:
            words = file.read().splitlines()
            if file_name == 'positive-words':
                positive_words = words
            elif file_name == 'negative-words':
                negative_words = words
```

```python
from bs4 import BeautifulSoup
import requests
import urllib3
import re

# Disable InsecureRequestWarning
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

def pipeline(df, row_num):
    url = df.at[row_num, 'URL']

    # Define proxies as a list
    proxies = ['18.230.145.231:3128', '13.48.26.64:8080', '3.16.22.38:4216']
    proxy = proxies[row_num % len(proxies)]  # Choose a proxy for each row

    print(f"Using the proxy: {proxy}")

    try:
        # Make the request using the current proxy
        res = requests.get(url, verify=False)

        if res.status_code == 200:
            html_content = res.text

            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(html_content, 'html.parser')

            # Find all <p> tags in the parsed HTML
            paragraphs = soup.find_all('p')

            # String to store extracted paragraphs
            extracted_text = ""

            # Extract text from all <p> tags and concatenate to the string
            for para in paragraphs:
                scraped = para.text
                if 'Contact us: hello@blackcoffer.com' in scraped:
                    scraped = scraped.split('Contact us: hello@blackcoffer.com')[0].strip()
                if '© All Right Reserved, Blackcoffer(OPC) Pvt. Ltd' in scraped:
                    scraped = scraped.split('© All Right Reserved, Blackcoffer(OPC) Pvt. Ltd')[0].strip()
                scraped = scraped.rstrip()
                extracted_text += scraped + " "  # Adding a space after each paragraph

            # Cleaning extracted text
            extracted_text = re.sub(r'\n+', '', extracted_text)
            extracted_text = re.sub(r'•', '', extracted_text)
            extracted_text = extracted_text.strip()

            words_string = re.sub(r'[.,!\'\’-]', '', extracted_text).split()

            common_words = [word for word in words_string if word in positive_words or word in negative_words]

            sentences = extracted_text.split('.')
            pos, neg = 0, 0
            words_c = len(words_string)

            for word in words_string:
                if word in positive_words:
                    pos += 1
                elif word in negative_words:
                    neg += 1

            df.at[row_num, 'POSITIVE SCORE'] = pos
            df.at[row_num, 'NEGATIVE SCORE'] = neg

            polarity_score = (pos - neg) / ((pos + neg) + 0.000001)
            subjectivity_score = (pos + neg) / (words_c + 0.000001)

            df.at[row_num, 'POLARITY SCORE'] = polarity_score
            df.at[row_num, 'SUBJECTIVITY SCORE'] = subjectivity_score

            total_syllables = 0

            def count_syllables(word):
                exceptions = ["es", "ed"]
                if word.endswith("es") or word.endswith("ed"):
                    return 0

                vowels = "aeiouAEIOU"
                count = 0

                for char in word:
                    if char in vowels:
                        count+=1
                return count

            complex_words = [word for word in words_string if count_syllables(word) > 2]

            avg_sent_len = len(set(words_string))/len(sentences)
            percent_complex = len(complex_words)/len(set(words_string))
            fog_index = 0.4*(avg_sent_len+percent_complex)

            df.at[row_num,'AVG SENTENCE LENGTH'] = avg_sent_len
            df.at[row_num,'PERCENTAGE OF COMPLEX WORDS'] = percent_complex
            df.at[row_num,'FOG INDEX'] = fog_index
            df.at[row_num,'AVG NUMBER OF WORDS PER SENTENCE'] = avg_sent_len

            df.at[row_num,'COMPLEX WORD COUNT'] = len(complex_words)
            df.at[row_num,'WORD COUNT'] = len(set(words_string))

#             Calculating syllables per word
            syllables_per_word = [count_syllables(word) for word in words_string]
            df.at[row_num, 'SYLLABLE PER WORD'] = sum(syllables_per_word) / len(words_string)

#             df.at[row_num,'SYLLABLE PER WORD'] = total_syllables/len(words_string)

            personal_pronouns = ['I','we','my','ours','us']
            personal_count=0
            for word in words_string:
                if word=='I' or word.lower() in personal_pronouns and word!='US':
                    personal_count+=1

            df.at[row_num,'PERSONAL PRONOUNS'] = personal_count

            chars = 0
            for word in words_string:
                for ch in word:
                    chars+=1
            avg_word_len = chars/len(words_string)

            df.at[row_num,'AVG WORD LENGTH'] = avg_word_len

        else:
            print(f"Failed to fetch the webpage. Status code: {res.status_code}")

    except Exception as e:
        print(f"Failed: {e}")
```
